## First come First serve (FCFS)

```python
import pandas as pd 
def FCFS(): 
    start_time=[]
    x=int(input('Enter the no. of processes :'))
    process=[]
    for i in range(1,x+1):
        process.append('p'+str(i))
    for i in range(x):
        print('Enter Start Time of process',process[i],':')
        a=int(input())
        start_time.append(a)
    cpu_time=[]
    for i in range(x):
        print('Enter CPU Time of process',process[i],':')
        a=int(input())
        cpu_time.append(a)

    arrival_time=[]
    for i in range(x):
        print('Enter Arrival Time of process',process[i],':')
        a=int(input())
        arrival_time.append(a)

    finish_time=[]
    for i in range(0,x):
        finish_time.append(start_time[i]+cpu_time[i])


    Turn_around=[]
    for i in range(0,x):
        Turn_around.append(finish_time[i]-arrival_time[i])


    waiting_Time=[]
    for i in range(0,x):
        waiting_Time.append(Turn_around[i]-cpu_time[i])

    avg_tat=0
    avg_wait=0
    for i in range(0,x):
        avg_tat=avg_tat+Turn_around[i]
        avg_wait=avg_wait+waiting_Time[i]
    df=pd.DataFrame(list(zip(process,start_time,finish_time,
                         cpu_time,arrival_time,Turn_around,waiting_Time)),
                columns=['Process','Start Time','Finish Time',
                         'CPU Time','Arrival Time','Turn Around Time','Waiting Time'])
    print('______________________________________________________________\n')
    print('Average Turn-Around Time :',avg_tat/x)
    print("Average Waiting Time :",avg_wait/x)
    return df
df=FCFS()   
df
```

# Shortest Job Next (SJN) 

> The Process having the less CPU time will be executed first

```python
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt 
import numpy as np


def SJN(): 
    start_time=[]
    x=int(input('Enter the no. of processes :'))
    process=[]
    for i in range(1,x+1):
        process.append('p'+str(i))
    for i in range(x):
        print('Enter Start Time of process',process[i],':')
        a=int(input())
        start_time.append(a)
    cpu_time=[]
    for i in range(x):
        print('Enter CPU Time of process',process[i],':')
        a=int(input())
        cpu_time.append(a)

    arrival_time=[]
    for i in range(x):
        print('Enter Arrival Time of process',process[i],':')
        a=int(input())
        arrival_time.append(a)

    finish_time=[]
    for i in range(0,x):
        finish_time.append(start_time[i]+cpu_time[i])


    Turn_around=[]
    for i in range(0,x):
        Turn_around.append(finish_time[i]-arrival_time[i])


    waiting_Time=[]
    for i in range(0,x):
        waiting_Time.append(Turn_around[i]-cpu_time[i])

    avg_tat=0
    avg_wait=0
    for i in range(0,x):
        avg_tat=avg_tat+Turn_around[i]
        avg_wait=avg_wait+waiting_Time[i]
    df=pd.DataFrame(list(zip(process,start_time,finish_time,
                         cpu_time,arrival_time,Turn_around,waiting_Time)),
                columns=['Process','Start Time','Finish Time',
                         'CPU Time','Arrival Time','Turn Around Time','Waiting Time'])
    print('______________________________________________________________\n')
    print('Average Turn-Around Time :',avg_tat/x)
    print("Average Waiting Time :",avg_wait/x)
    return df
df=SJN()   
df
```

```python
df.plot('Process','CPU Time',kind='barh',figsize=(12,4));
```

> Setting the index according ascending order of CPU Time

```python
l1=np.sort(df['CPU Time'])
index=[]
for i in range(0,len(l1)):
    for j in range(0,len(l1)):
        if l1[i]==df['CPU Time'][j]:
              index.append(j)
```

> Assigning the ascending index to DataFrame

```python
df=df.reindex(index)
```

> Plot Showing the Shortest Job will be Executed First

```python
df.plot('Process','CPU Time',kind='barh',figsize=(12,4));
```

## Priority Scheduling

```python
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt 
import numpy as np


def PS(): 
    start_time=[]
    x=int(input('Enter the no. of processes :'))
    process=[]
    for i in range(1,x+1):
        process.append('p'+str(i))
    for i in range(x):
        print('Enter Start Time of process',process[i],':')
        a=int(input())
        start_time.append(a)
    cpu_time=[]
    for i in range(x):
        print('Enter CPU Time of process',process[i],':')
        a=int(input())
        cpu_time.append(a)

    arrival_time=[]
    for i in range(x):
        print('Enter Arrival Time of process',process[i],':')
        a=int(input())
        arrival_time.append(a)
    
    priority=[]
    for i in range(x):
        print('Enter Priority of process',process[i],':')
        a=int(input())
        priority.append(a)

    finish_time=[]
    for i in range(0,x):
        finish_time.append(start_time[i]+cpu_time[i])


    Turn_around=[]
    for i in range(0,x):
        Turn_around.append(finish_time[i]-arrival_time[i])


    waiting_Time=[]
    for i in range(0,x):
        waiting_Time.append(Turn_around[i]-cpu_time[i])
        

    
    avg_tat=0
    avg_wait=0
    for i in range(0,x):
        avg_tat=avg_tat+Turn_around[i]
        avg_wait=avg_wait+waiting_Time[i]
        
    df=pd.DataFrame(list(zip(process,priority,start_time,finish_time,
                         cpu_time,arrival_time,Turn_around,waiting_Time)),
                columns=['Process','Priority','Start Time','Finish Time',
                         'CPU Time','Arrival Time','Turn Around Time','Waiting Time'])
    print('______________________________________________________________\n')
    print('Average Turn-Around Time :',avg_tat/x)
    print("Average Waiting Time :",avg_wait/x)
    return df
df=PS()   
df
```

> Setting the index according ascending order of Priority

```python
l1=np.sort(df['Priority'])
index=[]
for i in range(0,len(l1)):
    for j in range(0,len(l1)):
        if l1[i]==df['Priority'][j]:
              index.append(j)
```

> Assigning the ascending index to DataFrame

```python
df=df.reindex(index)
```

## Round Robin (RR) Scheduling

```python
x=int(input('Enter the number of processes :'))

states=[]
for i in range(1,x+1):
    states.append('p'+str(i))
cpu_time=[]


for i in range(x):
    print('Enter the CPU Time for ',states[i],':')
    a=int(input())
    cpu_time.append(a)
q=int(input('Enter Value of Time quantum :'))

t=0
ready=[]
cnt=0
remain=cpu_time

print('______________________________________________________________________')
print('Order of processes going to be executed :')
print('______________________________________________________________________')

while(True):
        all_zeros=True
        
        for i in range(x):
            if remain[i]>q:
                all_zeros=False
                t+=q
                print('p'+str(i+1),end='==>')
                remain[i]-=q
            else:
                if remain[i]>0:
                            t+=remain[i]
                            print('p'+str(i+1),end='==>')
                remain[i]=0
        if(all_zeros==True):
            break
```
